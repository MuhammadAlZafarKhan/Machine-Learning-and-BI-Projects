```python
# Packahes

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df1 = pd.read_csv("telecom_churn.csv")
df1.head(5)
```

```python
le = LabelEncoder()
```

```python
# Encoding the catrgorical features
df1["state"] = le.fit_transform(df1["state"])
df1["international plan"] = le.fit_transform(df1["international plan"])
df1["voice mail plan"] = le.fit_transform(df1["voice mail plan"])
df1["churn"] = le.fit_transform(df1["churn"])
```

```python
df1.head(5)
```

```python
# Dropping irrelevant columns: Phone number

df1.drop("phone number", inplace = True, axis = 1)
df1.head(5)
```

```python
# Checking data types
df1.dtypes
```

```python
# Converting relevant datatypes
df1["state"] = df1["state"].astype("category")
df1["area code"] = df1["area code"].astype("category")
df1["international plan"] = df1["international plan"].astype("category")
df1["voice mail plan"] = df1["voice mail plan"].astype("category")
df1["churn"] = df1["churn"].astype("category")

df1.dtypes
```

```python
X = df1[["state", "account length", "area code", "international plan", "voice mail plan", "number vmail messages", "total day minutes",
        "total day calls", "total day charge", "total eve minutes", "total eve calls", "total eve charge", "total night minutes", 
        "total night calls", "total night charge", "total intl minutes", "total intl calls", "total intl charge", "customer service calls"]]
y = df1[["churn"]]
```

```python
# Train and test splits
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Model Building

from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier


lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_score = lr.score(X_test, y_test)
dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_score = dt.score(X_test, y_test)
svm = SVC()
svm.fit(X_train, y_train)
svm_score = svm.score(X_test, y_test)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_score = rf.score(X_test, y_test)

print("Logistic regression score: ", lr_score)
print("Decision tree score: ", dt_score)
print("Support vector machine score: ", svm_score)
print("Ranform forest score: ", rf_score)
```
