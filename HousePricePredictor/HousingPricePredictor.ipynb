```python
# Packages
import pandas as pd
```

```python
# Importing data
df = pd.read_csv("HousingPrice.csv", sep = ";")
df.head()
```

```python
# Checking datatypes
df.dtypes
```

```python
# Extracting the features that we want
df1 = df[["MSSubClass", "MSZoning", "LotFrontage", "LotArea", "LotShape", "LandContour", "Utilities",
          "PoolArea", "YrSold", "SaleCondition", "SalePrice"]]
df1.head()
```

```python
# Encoding some features
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df1["MSZoning"] = le.fit_transform(df1["MSZoning"])
df1["LotShape"] = le.fit_transform(df1["LotShape"])
df1["LandContour"] = le.fit_transform(df1["LandContour"])
df1["Utilities"] = le.fit_transform(df1["Utilities"])
df1["SaleCondition"] = le.fit_transform(df1["SaleCondition"])
```

```python
# Converting the relevant features to categorical type, and the "SalePrice" is a float
df1["MSSubClass"] = df1["MSSubClass"].astype("category")
df1["MSZoning"] = df1["MSZoning"].astype("category")
df1["LotShape"] = df1["LotShape"].astype("category")
df1["LandContour"] = df1["LandContour"].astype("category")
df1["Utilities"] = df1["Utilities"].astype("category")
df1["YrSold"] = df1["YrSold"].astype("category")
df1["SaleCondition"] = df1["SaleCondition"].astype("category")
df1["SalePrice"] = df1["SalePrice"].astype("float64")
```

```python
# Checking datatypes to ensure proper encoding
df1.dtypes
```

```python
# Checking for null values
df1.isnull().sum()
```

We observe **259** null values from the variable "LotFrontage", we will simply ignore this variable in our subsequent
analysis, as since we have less data, dropping it will make us lose a lot of the data

```python
# Having a broad overview of our data
df1
```

```python
# Splitting the data into train and test sets
from sklearn.model_selection import train_test_split

X = df1[["MSSubClass", "MSZoning", "LotArea", "LotShape", "LandContour", "Utilities", "PoolArea", 
         "YrSold", "SaleCondition"]]
y = df1[["SalePrice"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

```python
# Building the model
from sklearn import linear_model

# Lasso (L1 regularisation) regression
lasso_reg = linear_model.Lasso(alpha = 50, max_iter = 1000, tol = 0.1)
lasso_reg.fit(X_train, y_train)
lasso_reg.score(X_test, y_test)
```

```python
# Ridge (L2 regularisation) regression
ridge_reg = linear_model.Ridge(alpha = 50, max_iter = 1000, tol = 0.1)
ridge_reg.fit(X_train, y_train)
ridge_reg.score(X_test, y_test)
```

Clearly, these low scores are indicative of real-world data versus working with fictitious data. Also, we can improve the score by adding more observations!
